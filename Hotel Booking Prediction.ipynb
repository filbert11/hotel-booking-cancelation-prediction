## Remove unwanted column

```python
data.drop(columns=['ReservationStatus','ReservationStatusDate','ArrivalDateMonth','ArrivalDateDayOfMonth','ArrivalDateYear','RequiredCarParkingSpaces',], inplace=True)
data['ArrivalDateWeekNumber']= data['ArrivalDateWeekNumber'].astype(object)
categorical = [var for var in data.columns if data[var].dtype=='O']
numerical = [var for var in data.columns if data[var].dtype!='O']


#global remove not relavent columns
numerical.remove('IsCanceled')
print(numerical)
print(categorical)

#Filter to only hotel room cheaper than 0 - 1000
data = data[data['ADR']<1000]
data = data[data['ADR']>0]
data = data[data['Adults']<=4]
data = data[data['StaysInWeekNights']<=9]
data = data[data['Babies']<=9]
data = data[data['Children']<=9]

#drop no data children, country
data = data[data.Children.notnull()]
data = data[data.Country.notnull()]
```

### Check for Collinearity

```python
# EDA

def plot_corr(df,size):
    '''
    Function plots graphical corelation matrix for each pair of column in dataframe
    Input: 
          df: pandas dataframe
          size: vertical and horizontal size of the plot
    Displays: 
            matrix of corelation between columns. Blue-cyan-yellow-red-darkred => less to more corelated
                                                  0-------------------->1
                                                  Expect a dark red line running from top left to bottom right
    '''

    corr=df.corr()  #data frame corelation funnction
    fig,ax= plt.subplots(figsize=(size,size))
    ax.matshow(corr)  #color code the matrix rectangles by corelation value
    plt.yticks(range(len(corr.columns)),corr.columns) #draw xticks mark
    plt.xticks(range(len(corr.columns)),corr.columns,rotation=90) #draw xticks mark

plot_corr(data[numerical],20)
```
